```python
#Chaitali Chitnis, BIOE442

%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#Import dataset with correctly labeled columns for your dataframe

df = pd.read_csv('wdbc.data')

features = ['radius','texture','perimeter','area','smoothness','compactness','concavity','concave_pts','symmetry','fractal_dim']
column_names = ['ID','Diagnosis']
for i in range(10):
    column_names.append(features[i]+'_mean')
    column_names.append(features[i]+'_se')
    column_names.append(features[i]+'_worst')

df.columns = column_names
```

```python
#Output first several lines of dataframe

df.head()
```

```python
#Produce one plot from the data

f,ax = plt.subplots()
ax.scatter(df.radius_mean, df.symmetry_mean, label = 'o')
ax.set_xlabel('Mean Radius')
ax.set_ylabel('Mean Symmetry Score')
ax.set_title('Mean Symmetry Score v Mean Radius')
plt.show()

#I chose to plot symmetry against radius to see how a change in size affected the overall symmetry of the malignancy
```
